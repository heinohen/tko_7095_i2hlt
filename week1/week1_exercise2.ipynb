Extract json to a list of dicts

```python
import json


data = []
with open('topic1.jsonl') as f:
  for l in f:
    d = json.loads(l)
    data.append(d)


data2 = []
with open('topic2.jsonl') as f:
  for r in f:
    e = json.loads(r)
    data2.append(e)
```

Join dicts as one string

```python

lauseet1 = [dicti['text'] for dicti in data]
sanajono1 = ' '.join(lauseet1)

lauseet2 = [dicti['text'] for dicti in data2]
sanajono2 = ' '.join(lauseet2)

```

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud


def show_word_cloud(text):
  cloud = WordCloud(width=1000, height=1000, background_color='white')
  cloud = cloud.generate(text)
  plt.figure(figsize=(10,10))
  plt.imshow(cloud)
  plt.axis('off')
```

Generate wordcloud from topic1. Looks like it is about elections and politics

```python

show_word_cloud(sanajono1)
```

Generate wordcloud from topic2. Looks like its about weather forecast.

```python
show_word_cloud(sanajono2)
```
