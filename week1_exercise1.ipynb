```python
from conllu import parse_incr

lemmat = dict()

data_file = open("fi_tdt-ud-train.conllu", "r", encoding="utf-8")
for tokenlist in parse_incr(data_file): # from https://pypi.org/project/conllu/

    for token in tokenlist:
      if token['upos'] == 'VERB': # attributes from https://universaldependencies.org/format.html
        if token['lemma'] in lemmat:
            lemmat[token['lemma']].add(token['form'])
        else:
            lemmat[token['lemma']] = {token['form']}

sum = 0
longest = 0
longest_lemma = ""
for k,v in lemmat.items():
  sum += len(v)
  if len(v) > longest:
    longest = len(v)
    longest_lemma = k
  print(k,v, len(v))

print('They always say a Finnish verb has 10,000 different forms. How many of these do you see in reality, i.e. in this dataset? ')
print(f'average {sum / len(lemmat)}, amount of most different forms is for \"{longest_lemma}\" with {longest} different forms')
```
